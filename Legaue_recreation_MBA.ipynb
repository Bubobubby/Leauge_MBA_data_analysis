```python
r=requests.get("https://hosted.dcd.shared.geniussports.com/LMBA/en/competition/35420/team/159891/statistics")

soup = bs(r.content,'lxml')

contents=soup.prettify()
#contents
```

```python
### zbiorcze staty White Eagles

table_total_html=soup.select('table',{'class':'tableClass footable team-stats footable-done footable-1 breakpoint-md'})[0]
table_total_columns=table_total_html.find('thead').find_all('th')
table_total_columns_names=[c.string for c in table_total_columns]

table_total_rows=table_total_html.find('tbody').find_all('tr')
table_total_list=[]
for tr in table_total_rows:
	td=tr.find_all('td')
	row=[tr.get_text().replace("\n"," ") for tr in td]
	table_total_list.append(row)

table_total=pd.DataFrame(table_total_list,columns=table_total_columns_names)
table_total
```

```python
#### Statystyki z każdego meczu- IN PROGRESS

base_path="https://hosted.dcd.shared.geniussports.com/LMBA/en/competition/35420/match/"

num_game='2233435'
end_path='/boxscore?'

full_path=base_path+num_game+end_path

def get_info_box(url,numberbox):

	r=requests.get(url)
	soup=bs(r.content,'lxml')
	info_box_host=soup.select('table',{'class':'tableClass footable footable-done footable-1 breakpoint-lg'})[numberbox]
	info_columns=info_box_host.find('thead').find_all('th')
	info_rows=info_box_host.find('tbody').find_all('tr')
	columns_names=[c.string for c in info_columns]
	team_table=[]
	for tr in info_rows:
		td=tr.find_all('td')
		row=[tr.get_text("",strip=True).replace("\n"," ") for tr in td]
		team_table.append(row)

	box_team_table=pd.DataFrame(team_table,columns=columns_names)
	box_team_table

get_info_box(full_path,1)
```
